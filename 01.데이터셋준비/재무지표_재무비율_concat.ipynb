```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_index = pd.read_csv("../data/processed_data/재무지표_150개이상제거완.csv", index_col=0)
# df_index.head()
```

```python
df_ratio = pd.read_csv("../data/processed_data/재무비율_처리완.csv")
# df_ratio.head()
```

```python
print(len(df_index.columns))
print(len(df_ratio.columns))
print('-------------------')
print(len(df_index))
print(len(df_ratio))
```

```python
df_index[df_index.columns[9:]] = df_index[df_index.columns[9:]].fillna(0)

df_index.columns[df_index.isnull().sum() > 0]
```

```python
df_merged = pd.merge(df_index, df_ratio,
                     on = ['회사명', '거래소코드', '회계년도', '소속코드', '통계청 한국표준산업분류 코드 10차(대분류)', '산업코드', '산업명', '상장일', '상장폐지일'],
                     how='inner', suffixes=('_index', '_ratio'))
df_merged.drop(columns = 'label_index', inplace = True)
df_merged.rename(columns = {'label_ratio' : 'label'}, inplace = True)
```
