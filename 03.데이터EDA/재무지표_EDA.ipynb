## 윈저라이징 전 1(분식)기업과 0(정상)기업의 기초통계량

```python
df[df['label']==1].describe()
```

```python
df[df['label']==0].describe()
```

```python
def winsorize_series(series, lower_percentile=0.01, upper_percentile=0.99):
    lower_bound = series.quantile(lower_percentile)
    upper_bound = series.quantile(upper_percentile)
    return series.clip(lower=lower_bound, upper=upper_bound)

# 9번째 열부터 끝까지 윈저라이징 적용
df_winsorized = df.iloc[:, 9:].apply(winsorize_series)

# 원래 데이터프레임의 9번째 열 이후 부분만 윈저라이징한 결과로 교체
df.iloc[:, 9:] = df_winsorized

# 결과 출력
print(df)
```
