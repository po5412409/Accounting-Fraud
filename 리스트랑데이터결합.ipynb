### 1. kospi, kosdaq 데이터셋 결합
코스피 8676 + 코스닥 18974 = 상장 27650

```python
import pandas as pd
```

```python
kospi = pd.read_csv('./data/코스피_재무비율.csv', encoding='cp949')
kosdaq = pd.read_csv('./data/코스닥_재무비율.csv', encoding='cp949')
```

```python
# 행기준으로 코스피, 코스닥 데이터 셋 결합
df = pd.concat([kospi, kosdaq], axis = 0)

# 인덱스 다시 0부터 27650까지
df.reset_index(drop = True, inplace = True)
df.tail(3)
```

### 2. 분식기업 리스트(list)와 코스피코스닥 원본 재무비율 데이터 결합을 위해 전처리

#### list
1. 컬럼명 변경__기업코드 -> 거래소코드, 결산기 -> 결산기
2. 데이터타입 변경__거래소코드(숫자형) -> (문자형)
3. 컬럼 삭제__유형, 유형.1, 유형.2, 유형.3, 유형.4

```python
list = pd.read_csv('./data/분식기업리스트.csv')
list
```

```python
# 1. 컬럼명 변경
list.rename({'기업코드' : '거래소코드', '결산기' : '결산기'}, inplace=True, axis = 1)

# 2. 데이터타입 변경
list['거래소코드'] = list['거래소코드'].astype(str)

# 소수점(.0)을 제거하고 정수 형태의 문자열로 변환 (4800.0 -> '4800')
list['거래소코드'] = list['거래소코드'].str.split('.').str[0]

# 3. 컬럼 삭제
list.drop(columns = {'유형', '유형.1', '유형.2', '유형.3', '유형.4'}, inplace = True)
list
```

#### df(kospi + kosdaq)
1. 데이터타입 변경__거래소코드(숫자형) -> (문자형)
2. 컬럼명 변경__회계년도 -> 결산기
3. 데이터 수정__결산기 년도만 나오도록 2012/12 -> 2012

```python
# 1. 데이터 타입 변경
df['거래소코드'] = df['거래소코드'].astype(str)

# 2. 컬럼명 변경
df.rename({'회계년도' : '결산기'}, axis = 1, inplace = True)

# 3. 데이터 수정__list랑 merge할 때 결산기 데이터타입을 통일시키기 위해 int로 통일
df['결산기'] = pd.to_datetime(df['결산기'], format = '%Y/%m')
df['결산기'] = df['결산기'].dt.year
df.head(3)
```

### 3. list랑 df 결합

```python
af = pd.merge(list, df, on = ['거래소코드', '결산기'], how = 'inner')
af
```

### 4. feature 데이터 컬럼명 전처리

```python
def clean_column_name(col_name):
    col_name = col_name.replace('(*)', '').replace('(원)', '').replace('(백만원)', '')
    col_name = col_name.replace('[제조]', '')
    col_name = col_name.replace('(천원)', '')
    col_name = col_name.replace('(IFRS연결)', '')
    col_name = col_name.replace('.1', '')

    return col_name.strip()  


af.columns = [clean_column_name(col) for col in af.columns]
af
```
