```python
import pandas as pd
```

```python
values = pd.read_csv('../data/processed_data/재무계정.csv')
ratio = pd.read_csv('../data/processed_data/재무비율.csv')
```

```python
df = pd.merge(values, ratio, on = ['거래소코드', '회계년도', '소속코드', '통계청 한국표준산업분류 코드 10차(대분류)',
                              '산업코드', '산업명', '상장일', '상장폐지일'])
```
